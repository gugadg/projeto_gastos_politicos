```python
import pandas as pd
```

***skiprows pula linha; delimiter em vez de , utilizamos o ; e o encoding para padrão latino***

```python
df = pd.read_csv('ceaps_dataset.csv', skiprows=1, delimiter=';', encoding='latin1')
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df['DATA']
```

```python
df['DATA'] = df['DATA'].str.replace('/','-')
```

```python
df['DATE'] = pd.to_datetime(df['DATA'])#, errors='coerce').isnull().sum()
```

```python
df[df["DATA"] == '31-10-200']
```

```python
df['DATA'] = df['DATA'].str.replace('206', '2016')
```

```python
df.dtypes
```

```python
df['DIA'] = df['DATE'].dt.day
```

```python
df.head()
```

```python
df['VALOR_REEMBOLSADO'] = df['VALOR_REEMBOLSADO'].str.replace(',','.')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['VALOR_REEMBOLSADO'] = pd.to_numeric(df['VALOR_REEMBOLSADO'])
```

```python
df.dtypes
```

```python
df.loc[df['VALOR_REEMBOLSADO'] > 100, 'MAIOR_QUE_100'] = 1
```

```python
df['MAIOR_QUE_100'] = 0
```

```python
df.columns
```

```python
df = df.filter(['ANO', 'MES','DIA', 'SENADOR', 'TIPO_DESPESA','FORNECEDOR', 'DOCUMENTO', 'DETALHAMENTO', 'DATA', 'VALOR_REEMBOLSADO', 'DATE', 'MAIOR_QUE_100'])
```

```python
mapper = {'Aluguel de imóveis para escritório político, compreendendo despesas concernentes a eles.':'Aluguel de imóveis', 'Aquisição de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos. ': 'Aquisição de material', 'Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar': 'Contratações', 'Locomoção, hospedagem, alimentação, combustíveis e lubrificantes': 'Logística', 'Passagens aéreas, aquáticas e terrestres nacionais': 'Passagens', 'Divulgação da atividade parlamentar': 'Divulgação', 'Serviços de Segurança Privada': 'Segurança'}

df['TIPO_DESPESA'] = df['TIPO_DESPESA'].map(mapper)
```
